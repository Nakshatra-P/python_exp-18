```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
df = pd.DataFrame({
    'Category': ['A','B','C','D','E'],
    'Values':[0,70,30,90,60],
    'Sales':np.random.randint(100,500,5),
    'Profit':np.random.randint(10,200,5)
})

display(df)
```

```python
#1. Area Plot
plt.figure()
plt.fill_between(df['Category'],df['Values'],color='green',alpha=0.5)
plt.title("Nakshatra")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.figure(figsize=(7,4))
plt.fill_between(df['Category'],df['Sales'],
                 color='skyblue',alpha=0.5,label='Sales')
plt.fill_between(df['Category'],df['Profit'],
                 color='orange',alpha=0.5,label='profit')
plt.title("Nakshatra")
plt.xlabel("Category")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
#2. Pie Chart
plt.figure()
plt.pie(df["Values"],labels=df['Category'], autopct='%1.2f%%')
plt.title("Nakshatra")
plt.show()
```

```python
#3. Donut Chart
plt.figure()
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%')
centre_circle = plt.Circle((0,0),0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Nakshatra")
plt.show()
```

```python
#4. Boxplot

plt.figure()
sns.boxplot(x=df['Values'])
plt.title("Nakshatra")
plt.show()
```

```python
#5. Heatmap (Correlation)

plt.figure()
corr = df[['Sales','Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Nakshatra")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df[['Values','Sales','Profit']].corr()
plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True)
plt.title("Nakshatra")
plt.show()
```

```python
#6. Bubble Plot (3 Variable Chart)

import matplotlib.pyplot as plt
plt.scatter(df['Sales'],df['Profit'],s=df['Values']*10)
plt.title("Nakshatra")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

sns.scatterplot(x='Sales', y='Profit',
                size='Values',        # bubble size
                hue='Values',         # color
                data=df,
                sizes=(50, 300),      # min-max size
                palette='viridis')

plt.title("Nakshatra")

plt.show()
```
